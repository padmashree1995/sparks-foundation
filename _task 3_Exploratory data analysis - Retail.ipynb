```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(5,7))
ax.scatter(data['Sales'],data['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,7))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=6)
plt.show()
```

```python
import matplotlib.pyplot as plt
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(5,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
t_states = data['State'].value_counts().nlargest(10)
t_states
```

```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','purple'],alpha=0.9,figsize=(3,4))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
ps = data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(3,6))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=data)
plt.legend()
plt.show()
```
